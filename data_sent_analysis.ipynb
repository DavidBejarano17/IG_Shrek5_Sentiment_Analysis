### 3.1- Sentiment Analysis by text and emojis

```python
#Using VADER library to analyze the text of the commentd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Manually add some words to VADER’s lexicon and improve the model
new_words = {
    "disapprove": -2.0, "thumbs down": -2.0, "bad emoji": -2.0,
    "laughing": 2.0, "love": 2.0, "crying": -1.5, "angry": -2.0, "dead": -1.5
}
sia.lexicon.update(new_words)

# Dictionary to detect the sentiment of the emoji
emoji_sentiment_dict = {
    "😂": "Positive", "🤣": "Positive", "😭": "Negative", "😊": "Positive", "❤️": "Positive",
    "😢": "Negative", "😡": "Negative", "😠": "Negative", "💔": "Negative",
    "😐": "Neutral", "🖕": "Negative", "👎": "Negative", "💀": "Negative", "🤮": "Negative", "🤢": "Negative"
}

# Function to analyze emoji sentiment separately using the tex column
#that was the original column, rather than the translated column to be applied below

def analyze_emoji_sentiment(text):
    if pd.isna(text) or not isinstance(text, str):  
        return "No Emojis", 0  #Is there no emojis return the value

    # Extract only emojis
    emojis = ''.join(char for char in text if emoji.is_emoji(char))
    
    if not emojis:
        return "No Emojis", 0  

    # Check sentiment of each emoji with our previous dictionary
    sentiments = [emoji_sentiment_dict.get(char, "Neutral") for char in emojis]
    
    if "Negative" in sentiments:
        return "Negative", -0.6  # NEGATIVE WEIGHT
    elif "Positive" in sentiments:
        return "Positive", 0.6   # POSITIVE WEIGHT
    else:
        return "Neutral", 0  

#Dictionary to store the results
results = {}

#More impact in emojis: Visualizing the data, was found some comments
#with a neutral text and a negative emoji, the model recognized as a neutral comment,
#to fix this issue, it was incremented their valu to consider it in the final score
EMOJI_WEIGHT = 1.5  

# Application of the sentiment analysis of the text using VADER, row by row, in the translated column
for i, row in tqdm(Shrek_dataset_TR.iterrows(), total=len(Shrek_dataset_TR)):
    user = row["username"]  # Identify user
    translated_text = str(row["translated_comments"]).strip()

    # VADER sentiment on translated comments
    text_sentiment = sia.polarity_scores(translated_text) if translated_text else {"compound": 0, "neg": 0, "neu": 1, "pos": 0}
    
    # Application of the Emoji sentiment analysis on original comments "text" column
    emoji_text = str(row["text"]).strip()
    emoji_sentiment, emoji_adjustment = analyze_emoji_sentiment(emoji_text) #Use of emojis function for analysis

    # Final compound score with increased the emoji weight for improve the analysis
    final_compound = text_sentiment["compound"] + (emoji_adjustment * EMOJI_WEIGHT)
    final_compound = max(-1, min(1, final_compound)) 

    # Store results in the dictionary
    results[user] = {
        "text_sentiment": text_sentiment["compound"],  # Store only compound score for clarity
        "emoji_sentiment": emoji_sentiment,
        "final_sentiment": final_compound
    }

# Convert the results dictionary into a DataFrame
df_sentiment = pd.DataFrame.from_dict(results, orient="index").reset_index().rename(columns={"index": "username"})

# Merge results back into the previous dataset
Shrek_dataset_TR = Shrek_dataset_TR.merge(df_sentiment, on="username", how="left")
```

### 3.2- Detection of the region by their langauge

```python
#Application of the assignation of region by detecting the language 
#Library for detecting the language
import langid

#Function that checks if text column contains only emojis
def is_only_emojis(text):
    return isinstance(text, str) and all(emoji.is_emoji(char) for char in text.strip())

#Function to detect language safely using langid
def detect_language(text):
    if is_only_emojis(text):  
        return "emoji"  #Assign a special label identificatioj for only emoji comments
    try:
        return langid.classify(text)[0]  # Detect language
    except:
        return "unknown"  # If detection fails, return "unknown"

# Function to map detected language to a general region
def map_language_to_region(lang):
    language_region_map = {
        "en": "North America / UK",
        "es": "Latin America / Spain",
        "fr": "France / Canada",
        "de": "Germany / Central Europe",
        "pt": "Brazil / Portugal",
        "it": "Italy",
        "ru": "Russia / Eastern Europe",
        "zh": "China",
        "ja": "Japan",
        "ko": "Korea",
        "ar": "Middle East",
        "hi": "India",
        "nl": "Netherlands",
        "unknown": "Other",
        "emoji": "Emoji-Based Region"  #Custom region for emoji-only comments
    }
    return language_region_map.get(lang, "Other")

#Apply `langid` language detection to the "text" column
Shrek_dataset_TR["detected_language"] = Shrek_dataset_TR["text"].apply(detect_language)

# Map detected languages to regions
Shrek_dataset_TR["region"] = Shrek_dataset_TR["detected_language"].apply(map_language_to_region)
```
